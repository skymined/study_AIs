```python
import pandas as pd
```

```python
import pyLDAvis
import pyLDAvis.lda_model
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.datasets import fetch_20newsgroups
# 데이터 로드
data = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'))
documents = data.data
```

```python
documents[:1]
```

### TfidfVectorizer
max_df : 반복이 너무 많은 문장들을 빼세요. 특징적인 어떤 것을 잡아낼 때 의미가 없다고 생각하는 것. 그것을 퍼센트나 개수로 지정
min_df : 최소 반복 숫자

```python
tfidfvectorizer = TfidfVectorizer(stop_words='english', max_df=0.95 , min_df=2 )
```

```python
tfidf = tfidfvectorizer.fit_transform(documents) # document는 list이기 때문에 그대로 넣어도 됨
```

```python
tfidf.toarray()[:1].shape  #39115차원
```

```python
import numpy as np
np.where(tfidf.toarray()[:1] > 0)  #where의 역할
```

```python
latentDirichletAllocation = LatentDirichletAllocation(n_components=5, n_jobs=-1) # 토픽 몇 개를 잡아낼 것인지 개수를 넣어줘야 함(n_components), n_jobs=-1은 무한대로!
```

```python
latentDirichletAllocation.fit(tfidf) # 문장 단위로 근접상황?
```

```python
vis = pyLDAvis.lda_model.prepare(latentDirichletAllocation,tfidf,tfidfvectorizer  ) #모델, 교육이 끝난 값, fit한 값
```

```python
pyLDAvis.enable_notebook()

pyLDAvis.display(vis) #반복된 것을 찾는 것이 아니라, 1을 이루는 문장 중에서 중요하다고 생각되는 요소들을 가중치를 두는 것
# 동그라미들이 떨어져 있을 수록 페르소나가 완벽한 것
```
