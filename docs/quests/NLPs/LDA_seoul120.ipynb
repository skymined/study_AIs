```python
import pandas as pd
import pyLDAvis #토픽 모델링 결과를 시각화하기 위한 것
import pyLDAvis.lda_model
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.datasets import fetch_20newsgroups
# 데이터 로드
data = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'))
documents = data.data
```

```python
documents[:1], type(documents)
```

### TfidfVectorizer
max_df / min_df

```python
tfidfVectorizer = TfidfVectorizer(stop_words='english', max_df=0.95, min_df=2)
```

```python
tfidf = tfidfVectorizer.fit_transform(documents)
```

```python
latentDirichletAllocation1 = LatentDirichletAllocation(n_components=5, n_jobs=-1)
```

```python
latentDirichletAllocation1.fit(tfidf)
```

### 1

```python
vis = pyLDAvis.lda_model.prepare(latentDirichletAllocation1, tfidf, tfidfVectorizer)
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.display(vis)
```

### 2

```python
latentDirichletAllocation2 = LatentDirichletAllocation(n_components=7, n_jobs=-1)
latentDirichletAllocation2.fit(tfidf)
vis2 = pyLDAvis.lda_model.prepare(latentDirichletAllocation2, tfidf, tfidfVectorizer)
pyLDAvis.display(vis2)
```
