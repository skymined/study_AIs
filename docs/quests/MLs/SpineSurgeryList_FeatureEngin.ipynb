### Scaling - MinMax
- 수치형 : 연령, 체중, 입원기간

```python
from sklearn.preprocessing import MinMaxScaler

def makescaling(dataframe, data) :
  minmax = MinMaxScaler()
  minmax.fit(dataframe[data].values.reshape(-1,1))
  result = minmax.transform(dataframe[data].values.reshape(-1,1))
  return result
```

```python
df_SSL['연령'].value_counts()
```

```python
df_SSL['연령_scaling'] = makescaling(df_SSL,'연령')
df_SSL['연령_scaling'].describe() # 확인해봤으니 나머지 수치형도 대입
```

```python
df_SSL['체중_scaling'] = makescaling(df_SSL,'체중')
df_SSL['입원기간_scaling'] = makescaling(df_SSL,'입원기간')
```

### Onehot Encoding
- 범주형 : 성별, 환자통증정도, 흡연여부, 과거수술횟수

```python
# unique를 위한 함수 제작해보기ㅋㅅㅋ
def knowunique(dataset, columns) :
  result = dataset[columns].unique()
  return result
```

```python
from sklearn.preprocessing import OneHotEncoder
def onehot(dataset, columns) :
  hothot = OneHotEncoder()
  hothot.fit(dataset[[columns]])
  hothot.categories_ # array 생성하겠지?
  hothot.get_feature_names_out() # 각각의 array에 이름을 생성해요
  encoder_hothot = hothot.transform(dataset[[columns]]).toarray()
  df_encoder_hothot = pd.DataFrame(data=encoder_hothot, columns=hothot.get_feature_names_out())
  # 이제 행으로 concat
  dataset = pd.concat((dataset.reset_index(drop=True), dataset.reset_index(drop=True)), axis=1)
  return dataset
```

```python
df_SSL = onehot(df_SSL, '성별')
```
